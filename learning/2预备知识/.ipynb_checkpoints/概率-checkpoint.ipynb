```python
import torch

x=torch.ones([10])
x
```

```python
from torch.distributions import multinomial
help(multinomial.Multinomial)
```

```python
samples=multinomial.Multinomial(1000,x).sample((10,))
cumsum=torch.cumsum(samples,0)
estimates=cumsum/cumsum.sum(axis=1,keepdim=True)
estimates
```

```python
import matplotlib.pyplot as plt
plt.plot(estimates)
```
